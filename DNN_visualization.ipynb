# Net Visualization

## torchviz.make_dot
适用于Pytorch的可视化包

```python
!pip install torchviz
import torch
import torchvision
from torchviz import make_dot
#from torch.autograd import Variable
```

```python
net = torchvision.models.resnet34()
y = net(torch.zeros(1, 3, 224, 224, dtype=torch.float, requires_grad=False))
```

```python
make_dot(y, params=dict(net.named_parameters()))
```

## hiddenlayer

[hiddenlayer](https://github.com/waleedka/hiddenlayer)是一个轻量级的工具包，可以用于可视化神经网络图和训练指标，适用于Pytorch, Tensorflow, Keras

```python
!pip install hiddenlayer
import torch
import torchvision
import hiddenlayer as hl
```

```python
net = torchvision.models.resnet34()
hl_graph = hl.build_graph(net, torch.zeros([1, 3, 224, 224]))
hl_graph.theme = hl.graph.THEMES["blue"].copy()  # Two options: basic and blue
#hl_graph.save('./resnet_hiddenlayer')
hl_graph
```

## Netron


```python
import torch
import torchvision
```

```python
net = torchvision.models.resnet34()
torch.onnx.export(net, torch.randn((1, 3, 224, 224)), 'resnet34.onnx')
```

此时当前路径下会生成一个"resnet34.onnx"文件，将其用Netron在线工具打开即可
1. 首先我们进入[Netron在线可视化官网](https://netron.app/)<img src="https://img-blog.csdnimg.cn/20210420162704518.png">

2. 我们点击**Open Model**打开刚才保存的resnet34.onnx文件
3. 此时我们得到了漂亮的可视化结果<img src="https://img-blog.csdnimg.cn/20210420162947802.png">
